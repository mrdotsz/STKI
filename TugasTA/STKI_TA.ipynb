```python
#to lower case
data = list(df["Abstrak"].astype(str).str.lower())
# data =  list(df["judul"].astype(str).str.lower())

kd = []
i = 1
for d in data:
    
    #StopWordRemover
    factory = StopWordRemoverFactory()
    stopword = factory.create_stop_word_remover()
    
    #Tokenize
    stop = nltk.tokenize.word_tokenize(stopword.remove(str(d)))
    
    #Stemmer
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    katadasar = stemmer.stem(str(stop))
    
    kd.append(katadasar)
    print(str(i) + str(','), end='')
    # print(d)
    i=i+1

no = 0
with open('dataset_dokumen_hasil.csv', 'w', newline='', encoding='utf-8-sig') as csvfile:
    spamwriter = csv.writer(csvfile)
    spamwriter.writerow(["teks"])
    for d in kd:
      spamwriter.writerow([kd[no]])
      no = no + 1
```

```python
dataAfter = pd.read_csv('dataset_dokumen_hasil.csv',sep="|")
```

```python
query = "Pemilihan proyek potensial dari paket lelang"


#to lower case
data = query.split(" ")

#Query Kata Dasar
querykd = []
for d in data:
    
    #StopWordRemover
    factory = StopWordRemoverFactory()
    stopword = factory.create_stop_word_remover()
    
    #Tokenize
    stop = nltk.tokenize.word_tokenize(stopword.remove(str(d.lower())))
    
    #Stemmer
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    katadasar = stemmer.stem(str(stop))
    
    querykd.append(katadasar)
    # print(str(i)+ str(','), end='')
    # print(d)
    # i=i+1

#TF - DF
termFrequency = []
dokumenfrequency = []
for index,term in enumerate(querykd):
  countDokumen = []
  countDokumenFrequency = 0
  for dokumen in dataAfter['teks']:
    count = 0
    for kata in dokumen.split(' '):
      if kata == term:
        count += 1
    countDokumen.append(count)
    
    if count > 0:
      countDokumenFrequency += 1
  termFrequency.append(countDokumen)
  dokumenfrequency.append(countDokumenFrequency)

#IDF + 1
idfSatu = []
jumlahDokumen = len(df)
for i in dokumenfrequency:
  idfSatu.append(log(jumlahDokumen / (i+1)))


#Pembobotan TFIDF (bisa di ganti COSIM)
weight = termFrequency[:]
for i in range(len(weight)):
   for j in range(len(weight[i])):
     weight[i][j] = termFrequency[i][j] * idfSatu[i]

#Menghitung total bobot dokumen
jumlahWeight = []
for i in range(len(weight[0])):
  jumlahWeight.append([i,0])

for i in range(len(weight)):
  for j in range(len(weight[i])):
    jumlahWeight[j][1] += weight[i][j]

#sorting bobot dokumen tertinggi = relevan
JumlahWeight = sorted(jumlahWeight, key=itemgetter(1), reverse=True)
REL = 0
for i in jumlahWeight:
  if i[1] != 0.0:
    REL+=1

#Hasil Pencarian
print("Berikut adalah 20 Abstrak Jurnal yang relevan dengan Query Anda dan tingkat Relevansinya : ")


for i in range(0,20):
  print(str(i+1) + str('. ') + str(df['Abstrak'][JumlahWeight[i][0]]) + " | "  + str(JumlahWeight[i][1]))
  # print(str(JumlahWeight[i][0]+2) + ',', end='')
```
